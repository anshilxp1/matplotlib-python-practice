# Box plot 

```python
# Box plot -visualization tool that  is used to detect the outliers in the dataset
```

```python
# Outliers are basicaly extremly high or extremly low value that affects the mean only
```

```python
# quantile=>is the term that divided the data into four equal parts and each part contain equal amount of data
```

```python
# q0 = 0%
# q1 = 25%
# q2 = 50%
# q3 = 75%
# q4 = 100%

# # upper_limit = q3+1.5(IQR)
# # lower_limit = q1 - 1.5(IQR)
# value outide these limits are considered outliers.

# IQR = q3-q1    (IQR means interquartile Range where your 50% data lies or distn b/w the q1 and q3 point)
```

```python
df
```

```python
plt.boxplot(df['tip']);
```

```python
# means outliers are present in data.
```

```python
# Trimming
# Capping / winsorization
```

```python
df['tip'].describe()
```

```python
q1=np.percentile(df['tip'],25)  
q1
```

```python
q3 = df['tip'].quantile(0.75)
q3
```

```python
IQR = q3-q1
IQR
```

```python
ul=q3+1.5*IQR
ul
```

```python
ll = q1-1.5*IQR
ll
```

```python
df[df['tip']>ul]
```

```python
df[df['tip']<ll]
```

```python
plt.boxplot(df['tip']);
```

```python
df
```

```python
df[df['tip']<ul]
```

```python
df['new_tip'] = np.where(df['tip']>ul , ul , np.where(df['tip']<ll,ll, df['tip']))  # here is nested loop


#np.where(conditon,value_if_true,value_if_false)
```

```python
df
```

```python
plt.boxplot(df['new_tip']); # now there is no outlier
```
